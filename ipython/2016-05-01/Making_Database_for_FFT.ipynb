```python
# labering data -> pandas dateframe
# freq -> sampling time [seconds]
# overlap -> % to freq
def sliding_window(labeling_data,freq, overlap):
    
    count =0
    start =labeling_data.Time[0]
    goal = start+datetime.timedelta(seconds=freq)
    add=freq*overlap
    yield labeling_data[(start <= labeling_data.Time) &(labeling_data.Time <= goal)]
       
    while True:
        try:
            count +=1
            start =start+datetime.timedelta(seconds=add)
            goal  =start+datetime.timedelta(seconds=freq)
            #print goal
            yield labeling_data[(start <= labeling_data.Time) &(labeling_data.Time <= goal)]
        except StopIteration:
            break
```

```python
def GetLabelNam(samp):
    l=['1st', '2nd', '3rd', '4th', '5th', '6th', '7th', '8th', '9th', '10th', '11th',
    '12th', '13th', '14th', '15th', '16th', '17th', '18th', '19th', '20th', '21st',
    '22nd', '23rd', '24th', '25th', '26th', '27th', '28th', '29th', '30th', '31st']
    
    return l[0:(samp/2)]
```

```python
start_time = T.clock()

name=['Cleaning-Conv-left-hand-ags.csv','Cleaning-Conv-left-leg-ags.csv','Cleaning-Conv-right-hand-ags.csv',\
     'Cleaning-Conv-right-leg-ags.csv','Cleaning-Conv-west-ags.csv']

sample =[16,32,64,128,256,512]

for n in name:
    for samp in sample:
        print n
        print "sample bit="+str(samp)
        
        Source=pd.read_csv(n,parse_dates=['Time'])
        freq=0.01*samp
        df=sliding_window(Source,freq,0.5)
        lAccX=np.array([])
        lAccY=np.array([])
        lAccZ=np.array([])
        lGyrX=np.array([])
        lGyrY=np.array([])
        lGyrZ=np.array([])
        time=[]
        label=np.array([])

        while True:
            try:
                d=df.next()
                # fftの直流成分を除くsample/2のみ
                lAccX=np.append(lAccX, np.array(fftpack.fft(d.AccX)[1:(samp/2)+1]) )
                lAccY=np.append(lAccY, np.array(fftpack.fft(d.AccY)[1:(samp/2)+1]) )
                lAccZ=np.append(lAccZ, np.array(fftpack.fft(d.AccZ)[1:(samp/2)+1]) )
                lGyrX=np.append(lGyrX, np.array(fftpack.fft(d.GyrX)[1:(samp/2)+1]) )
                lGyrY=np.append(lGyrY, np.array(fftpack.fft(d.GyrY)[1:(samp/2)+1]) )
                lGyrZ=np.append(lGyrZ, np.array(fftpack.fft(d.GyrZ)[1:(samp/2)+1]) )
                label=np.append(label, d.label.min())
                time=np.append(time, min(d.Time) )
            except ValueError:
                break
        
        time=np.array(time,dtype=np.datetime64)

        array =  GetLabelNam(samp)
        TimeArray=[]
        NumArray=[]
        LabelArray=[]

        for t in range(len(time)):
            for i in range(len(array)):
                TimeArray.append(str(t))
                LabelArray.append(label[t])

        for i in range(len(time)):
            for a in array:
                NumArray.append(a)

        TimeArray=np.array(TimeArray)
        NumArray=np.array(NumArray)
        LabelArray=np.array(LabelArray)
        
        if len(lAccX) < len(TimeArray):
            TimeArray = TimeArray[:len(lAccX)]
            NumArray = NumArray[:len(lAccX)]
            LabelArray = LabelArray[:len(lAccX)]
        elif len(lAccX) > len(TimeArray):
            lAccX=lAccX[:len(TimeArray)]
            lAccY=lAccY[:len(TimeArray)]
            lAccZ=lAccZ[:len(TimeArray)]
            lGyrX=lGyrX[:len(TimeArray)]
            lGyrY=lGyrY[:len(TimeArray)]
            lGyrZ=lGyrZ[:len(TimeArray)]

        AccX_real = np.real(lAccX)
        AccY_real = np.real(lAccY)
        AccZ_real = np.real(lAccZ)
        GyrX_real = np.real(lGyrX)
        GyrY_real = np.real(lGyrY)
        GyrZ_real = np.real(lGyrZ)
        AccX_imag = np.imag(lAccX)
        AccY_imag = np.imag(lAccY)
        AccZ_imag = np.imag(lAccZ)
        GyrX_imag = np.imag(lGyrX)
        GyrY_imag = np.imag(lGyrY)
        GyrZ_imag = np.imag(lGyrZ)
        
        fftdf=pd.DataFrame({
                'AccX_real':AccX_real[:len(TimeArray)],
                'AccY_real':AccY_real[:len(TimeArray)],
                'AccZ_real':AccZ_real[:len(TimeArray)],
                'GyrX_real':GyrX_real[:len(TimeArray)],
                'GyrY_real':GyrY_real[:len(TimeArray)],
                'GyrZ_real':GyrZ_real[:len(TimeArray)],
                'AccX_imag':AccX_imag[:len(TimeArray)],
                'AccY_imag':AccY_imag[:len(TimeArray)],
                'AccZ_imag':AccZ_imag[:len(TimeArray)],
                'GyrX_imag':GyrX_imag[:len(TimeArray)],
                'GyrY_imag':GyrY_imag[:len(TimeArray)],
                'GyrZ_imag':GyrZ_imag[:len(TimeArray)],
                'Time':TimeArray,
                'Number':NumArray,
                'Label':LabelArray
                }
              )

        #fftdf = fftdf.astype(np.float64)

        fftdf1=fftdf.set_index(['Time','Number'])
        filename= "FFT"+"-"+n+"-"+str(samp)+"bit"+".csv"

        fftdf1.to_csv(filename)
        
elapsed_time = T.clock() - start_time
print ("elapsed_time:{0}".format(elapsed_time)) + "[sec]"
```
