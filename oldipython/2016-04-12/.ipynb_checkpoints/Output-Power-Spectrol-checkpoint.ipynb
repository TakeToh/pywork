```python
data = pd.read_csv('Cleaning-Conv-left-hand-ags.csv')
df=data[data.label=='walking']
```

```python
freq, P=signal.welch(df.AccX, 100, nperseg=512)
```

```python
# output power spectrol

label = [u'walking',u'walking-turn',u'walking-back-arm',u'walking-back-arm-turn',\
        u'walking-cross-arm',u'walking-cross-arm-turn',u'stop']
data=[u'Cleaning-Conv-left-hand-ags.csv',u'Cleaning-Conv-left-leg-ags.csv',\
      u'Cleaning-Conv-right-hand-ags.csv',u'Cleaning-Conv-right-leg-ags.csv',\
      u'Cleaning-Conv-west-ags.csv']

for i in range(5):
    dataframe = pd.read_csv(data[i])
    df = dataframe[dataframe.label==label]
    print "*****"+data[i]+"*****"
    for col_name in ['AccX','AccY','AccZ','GyrX','GyrY','GyrZ']:
        print col_name
        freq, P=signal.welch(df.AccX, 100, nperseg=512)
        plt.xlabel("Frequency[Hz]")
        plt.xlim=[0,60]

        plt.ylabel("Power/frequency[dB/Hz]")
        plt.ylim=[0,10]

        filename=data[i]+'-'+col_name+'-'+label+'-powerspectrol'+'.png'
        plt.plot(freq, 10*np.log10(P), "r", linewidth=2, label=col_name)
        fig=plt.gcf()
        plt.show()
        print filename
        fig.savefig(filename,dpi=300)
```
