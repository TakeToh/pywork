```python
# input labeling_data freq->[s]
def sliding_window(data,label,freq):
    count =0
    labeling_data = raw_data[raw_data.label==label]
    yield labeling_data[\
    (labeling_data.Time[0] <= labeling_data.Time) &\
    (labeling_data.Time< labeling_data.Time[0]+datetime.timedelta(seconds=freq))]
    count +=1
    yield labeling_data[\
    (labeling_data.Time[0]+count*datetime.timedelta(seconds=freq) <= labeling_data.Time) &\
    (labeling_data.Time< labeling_data.Time[0]+(count+1)*datetime.timedelta(seconds=freq))]
```

```python
data=sliding_window(raw_data,5)
```
