```python
def plot_spectrogram(df,sampling_freq,xlim=[0,500],ylim=[0,30]):
    #plt.close('all')
    f,t,Sxx = signal.spectrogram(df,sampling_freq,nperseg=512)
    plt.xlabel(u"Time [sec]")
    plt.xlim(xlim)

    plt.ylabel(u"Frequency [Hz]")
    plt.ylim(ylim)

    plt.pcolormesh(t,f,Sxx)
    plt.colorbar()
    plt.show()
```

```python
def plot_spectrogram_columns(df,sampling_freq,xlim=[0,500],ylim=[0,30]):
    print "AccX spectrogram"
    plot_spectrogram(df.AccX,sampling_freq,xlim,ylim)
    print "AccY spectrogram"
    plot_spectrogram(df.AccY,sampling_freq,xlim,ylim)
    print "AccZ spectrogram"
    plot_spectrogram(df.AccZ,sampling_freq,xlim,ylim)
    print "GyrX spectrogram"
    plot_spectrogram(df.GyrX,sampling_freq,xlim,ylim)
    print "GyrY spectrogram"
    plot_spectrogram(df.GyrY,sampling_freq,xlim,ylim)
    print "GyrZ spectrogram"
    plot_spectrogram(df.GyrZ,sampling_freq,xlim,ylim)
```

```python
# output about walking

label=pd.read_csv('label-data.csv')
l = label['Info1'].drop_duplicates()

for i in range(5):
    for label in l:
        dataframe = pd.read_csv(data[i])
        df = dataframe[dataframe.label==label]
        print "*****"+data[i]+"*****"
        for col_name in ['AccX','AccY','AccZ','GyrX','GyrY','GyrZ']:
            print col_name
            f,t,Sxx = signal.spectrogram(df[col_name],100,nperseg=512)
            plt.xlabel(u"Time [sec]")
            plt.xlim=[0,60]

            plt.ylabel(u"Frequency [Hz]")
            plt.ylim=[0,10]

            plt.pcolormesh(t,f,Sxx)
            plt.colorbar()
            fig=plt.gcf()
            plt.show()
            filename=label+'/'+data[i]+'-'+col_name+'.png'
            print filename
            fig.savefig(filename,dpi=300)
```
