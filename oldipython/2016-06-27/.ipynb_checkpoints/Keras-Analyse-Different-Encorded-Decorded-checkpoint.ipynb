```python
def MakeGlaph(encordData,decordData,firstNum,goalNum,commonLabel):
    for i in range(0,encordData.itemsize):

        fig=plt.figure(figsize=(50,25))

        ax1=fig.add_subplot(221)
        ax2=fig.add_subplot(222)
        ax3=fig.add_subplot(223)
        ax4=fig.add_subplot(224)

        start = firstNum
        goal=start+1000
        saveName=str(start)+'-'+str(goal+3000)+'-glaph'

        dataE = encordData[start:goal,1]
        dataD = decordData[start:goal,1]
        ax1.plot(range(0,len(dataE)),dataE,label="Encoded")
        ax1.plot(range(0,len(dataD)),dataD,label="Decoded")
        ax1.legend(loc='best')
        ax1.set_title(str(start)+"-"+str(goal))


        dataE = encordData[start+1000:goal+1000,1]
        dataD = decordData[start+1000:goal+1000,1]
        ax2.plot(range(0,len(dataE)),dataE,label="Encoded")
        ax2.plot(range(0,len(dataD)),dataD,label="Decoded")
        ax2.legend(loc='best')
        ax2.set_title(str(start+1000)+"-"+str(goal+1000))

        dataE = s1encord[start+2000:goal+2000,1]
        dataD = s1decord[start+2000:goal+2000,1]
        ax3.plot(range(0,len(dataE)),dataE,label="Encoded")
        ax3.plot(range(0,len(dataD)),dataD,label="Decoded")
        ax3.legend(loc='best')
        ax3.set_title(str(start+2000)+"-"+str(goal+2000))

        dataE = encordData[start+3000:goal+3000,1]
        dataD = decordData[start+3000:goal+3000,1]
        ax4.plot(range(0,len(dataE)),dataE,label="Encoded")
        ax4.plot(range(0,len(dataD)),dataD,label="Decoded")
        ax4.legend(loc='best')
        ax4.set_title(str(start+3000)+"-"+str(goal+3000))

        plt.savefig(commonLabel+'-'+str(i)+'-'+saveName)
        plt.close('all')
```

```python
def CalcFFT(self,samp,overlap=0.5):
        start = time.time()
        fft_data = np.array([])
        
        for n,f in zip( self._columns,self._fft_col):
            print 'start'+n+'->'+f
            sw = self._sliding_window(n,samp,overlap)
            while True:
                try:
                    d=sw.next()
                    fft_data = np.append(fft_data, fftpack.fft(d)[1:(samp/2)+1] )   # fftの直流成分を除くsample/2の
                except StopIteration:
                    print 'CalcFFTStopIteration'
                    fft_data = fft_data.reshape(len(fft_data)/(samp/2),(samp/2) )
                    self._FFTData[f] = fft_data
                    np.savez(p_path+'fft/'+self.ClassName+'_'+str(f)+'_'+str(samp),self._FFTData[f])
                    break   
        elapsed_time = time.time() - start
        print ("elapsed_time:{0}".format(elapsed_time)) + "[sec]"
```

```python
Sensor ='sensor1'
encord=processing.LoadDicDataFromFileNPZ('study/'+Sensor+'_AccX_encoded.npz')
decord=processing.LoadDicDataFromFileNPZ('study/'+Sensor+'_AccX_decoded.npz')

fftDataE = np.array([])
fftDataD = np.array([])

for i in range(len(encord)):
        fftDataE=np.append(fftDataE, fftpack.fft(encord[i])[1:5])
        fftDataD=np.append(fftDataD, fftpack.fft(decord[i])[1:5])

fftDataE = fftDataE.reshape(len(encord),4)
fftDataD = fftDataD.reshape(len(encord),4)

for n in range(4000,16001,4000):
    MakeGlaph(np.imag(fftDataE), np.imag(fftDataD) ,n-4000,n,GlaphDataPath+Sensor+'-fft-image-AccX')
    
for n in range(4000,16001,4000):
    MakeGlaph(np.real(fftDataE), np.real(fftDataD) ,n-4000,n,GlaphDataPath+Sensor+'-fft-real-AccX')
    
PowerE = np.imag(fftDataE)**2+np.real(fftDataE)**2
PowerD = np.imag(fftDataD)**2+np.real(fftDataD)**2

for n in range(4000,16001,4000):
    MakeGlaph(PowerE, PowerD,n-4000,n,GlaphDataPath+Sensor+'-fft-power-AccX')
```
